```python
from PIL import Image

l = 90
r = 90
t = 100
b = 100

img = Image.open("car.png")
```

```python
img
```

```python
w, h = img.size
new_width = w + l + r
new_height = h + t + b
```

```python
result = Image.new(img.mode, (new_width,new_height), (0,0,0))
```

```python
result.paste(img,(l,t))
```
